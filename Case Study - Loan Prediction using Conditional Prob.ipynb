### Case Study: Predicting Loan Default Using Conditional Probability

#### Background

A financial institution, "CreditBank," wants to improve its loan approval process by predicting the likelihood of loan default based on customer profiles. The bank decides to use conditional probability to assess the risk associated with new loan applications.

#### Problem Definition

The primary question is: What is the probability that a customer will default on a loan given their financial profile?

#### Step 1: Collect Data

CreditBank collects and reviews historical data on past loan applicants, which includes:

*   Income levels
    
*   Employment status
    
*   Credit scores
    
*   Loan repayment histories
    

#### Step 2: Identify Key Factors

From the data, CreditBank identifies several key factors that historically correlate with the likelihood of default:

*   Low credit score
    
*   Unstable employment
    
*   High debt-to-income ratio
    

#### Step 3: Calculate Conditional Probabilities

CreditBank calculates the probability of default based on different conditions:

*   **P(Default | Low Credit Score)**
    
*   **P(Default | Unstable Employment)**
    
*   **P(Default | High Debt-to-Income Ratio)**

#### Step 5: Use the Model for Decision Making

CreditBank uses these probabilities to evaluate new loan applications. For instance, if an applicant has both a low credit score and unstable employment, the bank assesses the risk as higher compared to someone with stable employment and a high credit score.

#### Conclusion

Using conditional probability allows CreditBank to make informed decisions about loan approvals by quantifying the risk based on specific customer conditions. This approach helps minimize losses due to defaults while still providing loans to viable applicants.

This case study highlights how conditional probability can be applied in finance to assess risks and make decisions based on multiple factors of a customer's profile, demonstrating its value in business settings where risk assessment is crucial.
