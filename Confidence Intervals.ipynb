### Case Study: Confidence Intervals for Population Mean

**Objective**:To demonstrate how to construct and interpret confidence intervals for estimating the population mean of body temperatures using a sample.

**Background**:Confidence intervals provide a range of values which likely contain the population parameter (in this case, the mean) with a certain level of confidence (typically 95% or 99%). This case study will use a simulated dataset of body temperatures to construct a 95% confidence interval for the average body temperature.

### Methodology:

1.  **Simulate Data**:
    
    *   Generate a random sample of body temperatures normally distributed around a typical mean body temperature with a given standard deviation.
        
2.  **Calculate Sample Mean and Standard Error**:
    
    *   Calculate the mean and standard deviation of the sample.
        
    *   Determine the standard error of the mean by dividing the sample standard deviation by the square root of the sample size.
        
3.  **Construct Confidence Interval**:
    
    *   Using the sample mean and standard error, construct the confidence interval for the mean at a 95% confidence level.
        
4.  **Visualize the Results**:
    
    *   Plot the distribution of the sample and indicate the confidence interval on this distribution.
        

### Python Code Implementation

Let's implement this case study in Python:

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm




# Function to simulate body temperatures
def simulate_temperatures(n, mean=98.6, std_dev=0.7):
    return np.random.normal(mean, std_dev, n)

# Function to calculate confidence interval
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    margin = std_err * norm.ppf((1 + confidence) / 2)
    return mean - margin, mean + margin

# Simulate data
np.random.seed(0)
sample_size = 100
temperatures = simulate_temperatures(sample_size)

# Calculate confidence interval
ci_low, ci_high = confidence_interval(temperatures)

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(temperatures, bins=30, alpha=0.75, label='Sample Distribution')
plt.axvline(np.mean(temperatures), color='r', linestyle='dashed', linewidth=1, label='Sample Mean')
plt.axvline(ci_low, color='g', linestyle='dashed', linewidth=1, label='95% CI Lower')
plt.axvline(ci_high, color='g', linestyle='dashed', linewidth=1, label='95% CI Upper')
plt.title('Sample Distribution of Body Temperatures and Confidence Interval')
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### Analysis

*   **Sample Mean and CI**: The dashed red line represents the sample mean. The green lines represent the lower and upper bounds of the 95% confidence interval.
    
*   **Interpretation**: We are 95% confident that the interval from **ci\_low** to **ci\_high** contains the true average body temperature. This does not mean that 95% of individual temperatures fall within this range, but rather that the procedure used to construct the interval will capture the true population mean 95% of the time across many samples.
